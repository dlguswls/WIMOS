```python
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
```

```python
url = "https://ohou.se/store/category?affect_type=StoreHamburger&category=0"
driver = webdriver.Chrome('./chromedriver.exe')
driver.get(url)
```

```python
elem = driver.find_element_by_tag_name("body")
last_height = driver.execute_script("return document.body.scrollHeight")
```

```python
name_list = []
link_list = []
img_list = []
price_list = []
star_list = []
review_list = []
store_list = []
```

```python
while len(img_list)<=150:
    try : 
        #이미지 
        images = driver.find_elements_by_css_selector('div.col-lg-4>article>div.production-item-image>img')
        #상품 링크
        links = driver.find_elements_by_css_selector("div.col-lg-4>article> a")
        #회사명
        stores = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > h1 > span.production-item__header__brand")
        #제품명
        names = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > h1 > span.production-item__header__name")
        #가격
        prices = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > span.production-item-price > span.production-item-price__price")
        #별점
        stars = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > p > strong")
        #리뷰개수
        reviews = driver.find_elements_by_css_selector('div.col-lg-4>article>div.production-item__content > p')      
            
        if len(images) == 0:
            raise Exception
        

        for i in range(len(images)): 
            name_list.append(names[i].text)
            link_list.append(links[i].get_attribute('href'))
            img_list.append(images[i].get_attribute('src'))
            price_list.append(prices[i].text.rstrip(' 외'))
            star_list.append(stars[i].text)
            review_list.append(reviews[i].text.split()[2])
            store_list.append(stores[i].text)
            
    except Exception : 
        elem.send_keys(Keys.PAGE_DOWN)
        new_height = driver.execute_script("return document.body.scrollHeight")
        last_height = new_height
#         time.sleep(0.2)
        
    elem.send_keys(Keys.PAGE_DOWN)
    new_height = driver.execute_script("return document.body.scrollHeight")
    last_height = new_height
#     time.sleep(0.2)
```

```python
#이미지 
images = driver.find_elements_by_css_selector('div.col-lg-4>article>div.production-item-image>img')
#상품 링크
links = driver.find_elements_by_css_selector("div.col-lg-4>article> a")
#회사명
stores = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > h1 > span.production-item__header__brand")
#제품명
names = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > h1 > span.production-item__header__name")
#가격
prices = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > span.production-item-price > span.production-item-price__price")
#별점
stars = driver.find_elements_by_css_selector("div.col-lg-4>article>div.production-item__content > p > strong")
#리뷰개수
reviews = driver.find_elements_by_css_selector('div.col-lg-4>article>div.production-item__content > p')      

if len(images) == 0:
    raise Exception


for i in range(len(images)): 
    name_list.append(names[i].text)
    link_list.append(links[i].get_attribute('href'))
    img_list.append(images[i].get_attribute('src'))
    price_list.append(prices[i].text.rstrip(' 외'))
    star_list.append(stars[i].text)
    review_list.append(reviews[i].text.split()[2])
    store_list.append(stores[i].text)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
df['name'] = name_list
df['link'] = link_list
df['image'] = img_list
df['price'] = price_list
df['star'] = star_list
df['review'] = review_list
df['store'] = store_list
```
